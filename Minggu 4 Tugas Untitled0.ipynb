```python
# Tugas 1
# Fungsi untuk keluar dari program jika input gagal

print("-" * 30)
print("SELAMAT DATANG DI CAFE PLORA")
print("-" * 30)

# Menampilkan Menu (Sama seperti sebelumnya)
print("Menu Makanan:")
print("1. Ayam Bakar             - Rp 22.000")
print("2. Nasi Goreng            - Rp 15.000")
print("3. Kwetiau Goreng         - Rp 20.000")
print("4. Indomie Goreng         - Rp 12.000")
print("5. Mie Pangsit            - Rp 18.000")
print("-" * 30)
print("Menu Minuman & Dessert:")
print("6. Teh Obeng              - Rp 6.000")
print("7. Teh Tarik              - Rp 10.000")
print("8. Teh O                  - Rp 8.000")
print("9. Roti Bakar             - Rp 15.000")
print("10. Cookies               - Rp 12.000")
print("-" * 30)
print("-" * 30)

# Variabel utama
pilihan_str = input("Silakan masukkan nomor pilihan Anda (1-10): ")

# --- Bagian Pengecekan Menu (Sama seperti sebelumnya) ---
if pilihan == 1:
    nama_pesanan = "Ayam Bakar"
    harga_normal = 22000
elif pilihan == 2:
    nama_pesanan = "Nasi Goreng"
    harga_normal = 15000
elif pilihan == 3:
    nama_pesanan = "Kwetiau Goreng"
    harga_normal = 20000
elif pilihan == 4:
    nama_pesanan = "Indomie Goreng"
    harga_normal = 12000
elif pilihan == 5:
    nama_pesanan = "Mie Pangsit"
    harga_normal = 18000
elif pilihan == 6:
    nama_pesanan = "Teh Obeng"
    harga_normal = 6000
elif pilihan == 7:
    nama_pesanan = "Teh Tarik"
    harga_normal = 10000
elif pilihan == 8:
    nama_pesanan = "Teh O"
    harga_normal = 8000
elif pilihan == 9:
    nama_pesanan = "Roti Bakar"
    harga_normal = 15000
elif pilihan == 10:
    nama_pesanan = "Cookies "
    harga_normal = 12000
else:
    keluar_program("Maaf, pilihan menu tidak valid.")

print(f"Anda memesan: {nama_pesanan} dengan harga normal Rp {harga_normal:,.0f}")

# --- BAGIAN MODIFIKASI: LOGIKA DISKON ---
if harga_normal >= 50000:
    diskon_persen = 70
elif harga_normal >= 30000:
    diskon_persen = 50
elif harga_normal >= 20000:
    diskon_persen = 30

if diskon_persen > 0:
    diskon_nominal = int(harga_normal * (diskon_persen / 100))
    harga_setelah_diskon = harga_normal - diskon_nominal
else:
    harga_setelah_diskon = harga_normal

# --- Menampilkan detail diskon ---
if diskon_persen > 0:
    print(f"Selamat! Anda mendapatkan DISKON {diskon_persen}%")
    print(f"Diskon yang didapat   : Rp {diskon_nominal:,.0f}")
    print(f"Total Bayar           : Rp {harga_setelah_diskon:,.0f}")
    print("-" * 30)

# --- Proses Pembayaran ---
if harga_setelah_diskon > 0:
    print("-" * 30)
    # Penanganan kesalahan konversi input uang dibayar
    try:
        uang_dibayar_str = input(f"Total yang harus dibayar: Rp {harga_setelah_diskon:,.0f}. Masukkan jumlah uang Anda: Rp ")
        uang_dibayar = int(uang_dibayar_str)
    except ValueError:
        keluar_program("Input pembayaran tidak valid. Harap masukkan angka.")

    if uang_dibayar >= harga_setelah_diskon:
        kembalian = uang_dibayar - harga_setelah_diskon

        # --- Struk Pembayaran Akhir ---
        print("\n--- Struk Pembayaran ---")
        print(f"Pesanan Anda          : {nama_pesanan}")
        print(f"Harga Normal          : Rp {harga_normal:,.0f}")
        if diskon_persen > 0:
            print(f"Diskon ({diskon_persen}%)      : - Rp {diskon_nominal:,.0f}")
        print(f"Total Bayar           : Rp {harga_setelah_diskon:,.0f}")
        print(f"Uang Pembayaran       : Rp {uang_dibayar:,.0f}")
        print(f"Kembalian             : Rp {kembalian:,.0f}")
        print("------------------------")
        print("Terima kasih telah berbelanja!")
    else:
        kekurangan = harga_setelah_diskon - uang_dibayar
        print(f"Maaf, uang Anda kurang sebesar Rp {kekurangan:,.0f}.")

print("-" * 30)
```
