```python
# Soal 1
print("--- Soal 1 ---")
print("Bilangan Ganjil Antara 1-10")
# Mencetak bilangan ganjil 1 hingga 9
for angka in range(1, 10, 2):
    print(angka)
print("_"*40)

# Modifikasi Angka Genap Diantara 1-50
print("Modifikasi Angka Genap Diantara 1-50")
# Mencetak bilangan ganjil 1 hingga 9
for angka in range(2, 51, 2):
    print(angka)
```

```python
# Soal 2
print("--- Soal 2 ---")
print("Pemberhentian proses masukkan angka 99")
kode_berhenti = 99
input_pengguna = 0 # Inisialisasi awal

while input_pengguna > kode_berhenti:
    try:
        input_pengguna = int(input("Masukkan angka (99 untuk berhenti): "))
    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")

print("Perulangan telah dihentikan oleh sentinel (99).")
print("_"*60)

print("Modifikasi Dengan Menggunakan NIM")
kode_berhenti = 4232401030
input_pengguna = 0 # Inisialisasi awal

while input_pengguna != kode_berhenti:
    try:
        input_pengguna = int(input("Masukkan Nim (4232402030) untuk berhenti): "))
    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")
print("-"*60)
print("Perulangan telah dihentikan oleh sentinel Nim (4232402030).")
```

```python
# Soal 3
print("--- Soal 3 ---")
print("Menghitung total belanja")
harga_barang = [15000, 25000, 10000, 5000]
total_belanja = 0

for harga in harga_barang :
    total_belanja += harga
print("Total biaya belanja adalah:", total_belanja)
print("_"*40)

print("Modifikasi 10 item harga_barang")
harga_barang = [15000, 25000, 10000, 5000, 12000, 7000, 11000, 22000, 14000, 30000]
total_belanja = 0

for harga in harga_barang :
    total_belanja += harga

print("Total biaya belanja adalah:", total_belanja)
```

```python
# Soal 4
print("--- Soal 4 ---")
print("Batas percobaan")
percobaan_maksimal = 3
percobaan_saat_ini = 1

while percobaan_saat_ini <= percobaan_maksimal:
    print(f"Percobaan ke-{percobaan_saat_ini}...")
    # Asumsi ada logika input password di sini
    percobaan_saat_ini += 2

if percobaan_saat_ini == percobaan_maksimal + 1:
    print("Kesempatan habis.")
else:
    print("Akses diterima.")
print("_"*40)
print("Modifikasi Menggunakan password")
password_detail = "plorantina"
percobaan_maksimal = 3
percobaan_saat_ini = 1

while percobaan_saat_ini <= percobaan_maksimal:
    print(f"Percobaan ke-{percobaan_saat_ini}...")
    password = input("Masukkan password: ")

    if password == password_detail:
        print("Akses diterima.")
        break
    else:
        print("Password salah.\n")

    percobaan_saat_ini += 2
if percobaan_saat_ini > percobaan_maksimal:
    print("Kesempatan habis.")
```

```python
# Soal 5
print("--- Soal 5 ---")
print("Mencetak Karakter Teks")
nama_proyek = "PRAKTIKUM"
hasil_output = ""

for karakter in nama_proyek:
    hasil_output += karakter + "-"

print(hasil_output) # Output: P-R-A-K-T-I-K-U-M-
print("_"*50)

print("Modifikasi Menjadi Poloteknik Negeri Batam")
nama_proyek = "Politeknik Negeri Batam"
hasil_output = ""

for karakter in nama_proyek:
    hasil_output += karakter + "-"

print(hasil_output)
```

```python
# Soal 7
print("--- Soal 7 ---")
print("Mencetak Nama Bulan")
bulan = ["Januari", "Februari", "Maret", "April"]

for i, nama in enumerate(bulan):
    # i adalah indeks, nama adalah elemen
    print(f"Bulan ke-{i + 1}: {nama}")
print("_"*40)

print("Modifikasi Nama Kelurahan di Kota Batam")
kelurahan = [
    "Baloi Permai", "Batu Merah", "Buliang", "Kabil", "Sadai", "Tanjung Uma", "Teluk Tering", "Sei Panas", "Tembesi", "Belian"]

for i, nama in enumerate(kelurahan):
    print(f"Kelurahan ke-{i + 1}: {nama}")
```

```python
# Soal 8
print("--- Soal 8 ---")
print("Menggunakan Break")
data_angka = [10, 20, 30, -1, 50, 60]

for angka in data_angka:
    if angka < 0:
      print(f"angka negatif ditemukan{angka}")
      break
    print(angka)

print("Program berhenti karena angka -1 ditemukan.")
print("_"*50)

print("Modifikasi Program Berhenti Angka 0")
data_angka = [10, 20, 30, 0, 50, 60]

for angka in data_angka:
    if angka == 0:
        print(f"angka 0 ditemukan: {angka}")
        break
    print(angka)

print("Program berhenti karena angka 0 ditemukan.")
```

```python
# Soal 9
print("--- Soal 9 ---")
print("Penambahan Berkelanjutan dengan Batas")
total = 0
bilangan = 1
batas = 20

while total < batas:
    total += bilangan
    bilangan += 1

print("Total terakhir:", total)
print("_"*40)

print("Modifikasi")
total = 0
bilangan = 1
batas = 20

while total <= batas:
    total += bilangan
    bilangan += 1

# Menampilkan hasil akhir
print("Total terakhir:", total)
```
