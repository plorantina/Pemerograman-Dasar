Hasil penghitungan nilai unik (value_counts()) pada kolom jurusan mengubah semua menjadi huruf kecil dan menghilangkan spasi berlebih.
Kategori seperti 'energi' dan 'teknik' yang sebelumnya terpisah kini berhasil digabungkan ke dalam satu kategori baku (total energi menjadi 24 dan teknik menjadi 14).Namun, proses ini menyisakan sedikit pekerjaan rumah, yaitu menggabungkan varian 'manajemen energi' dan 'manajemenenergi' yang terpisah karena spasi, serta menentukan kategori untuk singkatan 'eng'.

```python
#untuk memudahkan replacing maka di buatkan mapping jurusan sesuai dengan yg di inginkan, dengan kata-kata yg mirip di masukan ke jurusan yg seusai

mapping_jurusan = {
    "energi": "Pembangkit Energi",
    " energi ": "Pembangkit Energi",
    "ENERGI": "Pembangkit Energi",
    "EnErGi": "Pembangkit Energi",
    "eng": "Pembangkit Energi",
    "teknik": "Pembangkit Energi",
    "Teknik ": "Pembangkit Energi",
    "teknik ": "Pembangkit Energi",
    "manajemenenergi": "Manajemen Energi",
    "manajemen energi": "Manajemen Energi"
}
```

mapping_jurusan bertujuan untuk menyeragamkan semua varian penulisan yang berbeda atau salah ketik agar jatuh ke dalam kategori jurusan yang benar.
(seperti 'teknik', 'energi', dan 'eng') menjadi kategori baku yang diinginkan (misalnya, "Pembangkit Energi" dan "Manajemen Energi").

```python
tasya_df["jurusan"] = tasya_df["jurusan"].replace(mapping_jurusan)
```

```python
tasya_df["jurusan"] = tasya_df["jurusan"].str.title()

#kemablikan menjadi huruf besar
```

mapping_jurusan untuk mendefinisikan penyeragaman. Misalnya, semua varian penulisan 'Energi' dan 'Teknik' diarahkan menjadi "Pembangkit Energi", dan varian 'manajemenenergi' diarahkan menjadi "Manajemen Energi".
Kode tasya_df["jurusan"].replace(mapping_jurusan) menerapkan penggantian massal sesuai mapping tersebut.
Kode (.str.title()) adalah mengembalikan format penulisan menjadi huruf besar di awal setiap kata.
